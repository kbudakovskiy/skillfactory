```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
churn_data = pd.read_csv('data/churn.csv')
churn_data.head()
```

```python
# Backup data
data_df = churn_data.copy()
```

## Crete graph showing loyal and exited clients rate, describe it 

```python
# Rate of exited and loyal customer
loyal_customer_rate = data_df.value_counts(['Exited'])
fig = plt.figure(figsize=(5, 5))
axes = fig.add_axes([0, 0, 1, 1])
loyal_clients_str = f'Loyal\n clients\n{loyal_customer_rate.loc[0].values[0]}'
exited_clients_str = f'Exited\nclients\n{loyal_customer_rate.loc[1].values[0]}'
total_account_str = f'Total clients account {loyal_customer_rate.loc[0].values[0] + loyal_customer_rate.loc[1].values[0]}'
axes.pie(
    loyal_customer_rate,
    labels=[loyal_clients_str,exited_clients_str],
    autopct='%.1f%%',
    explode = [0.1, 0]
);
```

```python
loyal_customer_rate.loc[0].values[0]
```

As you can see the majority of your clients, representing 79.6% of your customer base, are loyal. These clients consistently engage with your services, remain satisfied with their experience, and continue to support your business over time. Their loyalty indicates a strong level of trust, satisfaction, and value perception in your offerings.
A smaller portion of your clients (20.4 %),have exited or discontinued their relationship with your business. This could be due to various reasons such as dissatisfaction, switching to competitors, or changing needs. Understanding the reasons behind their departure is crucial for improving retention strategies and addressing any underlying issues.

```python
loyal_cliets_balance = data_df[data_df['Exited'] == 0]
exited_cliets_balance = data_df[data_df['Exited'] == 1]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 4))
sns.histplot(
    data=loyal_cliets_balance,
    x='Balance',
    bins=25,
    kde=True,
    ax=axes[0]
);

sns.histplot(
    data=exited_cliets_balance,
    x='Balance',
    bins=25,
    color='red',
    kde=True,
    ax=axes[1]
);
axes[0].set_title('Loyal client accounts balance', fontsize=12)
axes[1].set_title('Exited client accounts balance', fontsize=12)
```
